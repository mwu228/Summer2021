```python
#Preprocessing
dfEGAD6671_noNan_transpose = dfEGAD6671_noNan.T
dfEGAD6671_noNan_transpose.columns = geneList #Sets the column names to the gene SYMBOL, e.g. PTPN11
dfEGAD6671_noNan_transpose
```

```python
#Trimming the matrix by deleting GENE_ID, GENE_NAME, SYMBOL
dfEGAD6671_noNan_T_trim = dfEGAD6671_noNan_transpose.drop([0,1, 2]) #'GENE_NAME', 'SYMBOL'])
dfEGAD6671_noNan_T_trim
```

```python
#Gets rid of the very first column to make entire dataframe TPM values
dfEGAD6671_noNan_T_trim.reset_index(drop=True)
dfEGAD6671_noNan_T_trim
```

```python
#Forces any non-numeric values to be treated as float, to reduce errors with the corr() function later
dfEGAD6671_noNan_T_trim_numeric = pd.DataFrame()
dfEGAD6671_noNan_T_trim_numeric = dfEGAD6671_noNan_T_trim.apply(pd.to_numeric, errors='coerce')
dfEGAD6671_noNan_T_trim_numeric
```

```python
#Get rid of duplicate gene columns 
dfEGAD6671_noNan_T_trim_num_dup = dfEGAD6671_noNan_T_trim_numeric.loc[:,~dfEGAD6671_noNan_T_trim_numeric.columns.duplicated()]
dfEGAD6671_noNan_T_trim_num_dup
```

```python
#Define the cutoff for significant gene/protein expression
SIGEXPR = 0.45

#Create final dataframe to hold significant gene/protein expression
corrJak2Df = pd.DataFrame()
corrStat3Df = pd.DataFrame()
```

```python
#Calculate correlation for each column with respect to JAK2
#Will turn this into a function later
for col in geneList: 
    corrCoeff = dfEGAD6671_noNan_T_trim_num_dup["JAK2"].corr(dfEGAD6671_noNan_T_trim_num_dup[col])
    if(abs(corrCoeff) > SIGEXPR and col != "JAK2"):
        dfHolder = pd.DataFrame({col: corrCoeff}, index=[0])
        corrJak2Df = pd.concat([dfHolder, corrJak2Df], axis=1)
    
```

```python
#Calculate correlation for each column with respect to JAK2
#Will turn this into a function later
col = ''
for col in geneList: 
        corrCoeff2 = dfEGAD6671_noNan_T_trim_num_dup["STAT3"].corr(dfEGAD6671_noNan_T_trim_num_dup[col])
        print(corrCoeff2)
        if(abs(corrCoeff2) > SIGEXPR and col != "STAT3"):
            dfHolder2 = pd.DataFrame({col: corrCoeff2}, index=[0])
            corrStat3Df = pd.concat([dfHolder2, corrStat3Df], axis=1)
```
